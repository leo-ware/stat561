    a. Find the constant value of the p.d.f. on the rectangle.

4. **Suppose that X and Y have a continuous joint distribution for which the joint p.d.f. is defined as follows:**

$$
f(x, y) =
\begin{cases}
    cy^2 & \text{for } 0 \le x \le 2 \text{ and } 0 \le y \le 1 \\
    0 & \text{otherwise}
\end{cases}
$$


*a) Determine the value of $c$.*

$$
\int_0^2 \int_0^1 cy^2 dy dx = 1
$$

$$
\int_0^2 \int_0^1 y^2 dy dx = c^{-1}
$$

$$
\int_0^2 {(1)^3 \over 3} - {(0)^3 \over 3} dx = c^{-1}
$$

$$
\int_0^2 {1 \over 3} dx = c^{-1}
$$

$$
\int_0^2 dx = 3c^{-1}
$$

$$
2 = 3c^{-1}
$$

$$
c = {3 \over 2}
$$

*b) $P(X + Y > 2)$*

$$
P(X + Y > 2) =
\int_1^2 \int_{2 - x}^1 {3 \over 2}y^2 dy dx
$$

$$
P(X + Y > 2) =
{3 \over 2} \int_1^2 \int_{2 - x}^1 y^2 dy dx
$$

We use, $\int y^2 dy = {y^3 \over 3}$

$$
P(X + Y > 2) =
{3 \over 2} \int_1^2 {1 \over 3} - {(2 - x)^3 \over 3} dx
$$

$$
P(X + Y > 2) =
{1 \over 2} \int_1^2 1 - (2 - x)^3 dx
$$

Let $u = 2 - x$. Then, $du = -dx$, $x = 2 - u$.

$$
P(X + Y > 2) =
-{1 \over 2} \int_{2-1}^{2-2} 1 - u^3 du
$$

$$
P(X + Y > 2) =
{1 \over 2} \int_0^1 1 - u^3 du
$$

We use, $\int 1 - x^3 dx = x - {x^4 \over 4}$

$$
P(X + Y > 2) =
{1 \over 2} \left( 1 - {1 \over 4} \right) = {1 \over 2} {3 \over 4} = {3 \over 8}
$$

*c) $P(Y < 1/2)$*

The marginal $P(Y) = \int_0^2 {3 \over 2}y^2 dy = 3y^2$.

$P(Y < 1/2) = \int_0^{1/2} P(Y) dy = \int_0^{1/2} 3y^2 dy = (1/2)^3 = 1/8$

*d) $P(X \le 1)$*

The marginal $P(X) = \int_0^1 P(X, Y) dy = \int_0^1 {3 \over 2}y^2 dy = {1\over 2}$.

$P(X \le 1) = \int_0^1 P(X) dx = \int_0^1 {1 \over 2} dx = 1/2$

*e) $P(X = 3Y)$*

$P(X = 3Y) = 0$ because the area of a line is zero.

7. **Suppose that a point (X, Y) is to be chosen from the square S in the xy-plane containing all points (x, y) such that 0 ≤ x ≤ 1 and 0 ≤ y ≤ 1. Suppose that the probability that the chosen point will be the corner (0, 0) is 0.1, the probability that it will be the corner (1, 0) is 0.2, the probability that it will be the corner (0, 1) is 0.4, and the probability that it will be the corner (1, 1) is 0.1. Suppose also that if the chosen point is not one of the four corners of the square, then it will be an interior point of the square and will be chosen according to a constant p.d.f. over the interior of the square.**

*a) $P(X \le 1/4)$*

The probability that we will select a corner plot is $0.1 + 0.2 + 0.4 + 0.1 = 0.8$. So, the probability that we will select an interior point is $1 - 0.8 = 0.2$. The area of the square is 1. So, the constant value of the p.d.f. over the interior of the square is 0.2.

The $P(x \le 1/4)$ is the probability of $(0, 0)$ plus the probability of $(0, 1)$, plus the probability of getting an interior point such that $0 \le x \le 1/4$ and $0 \le y \le 1$.

The area of the interior region is $1/4 * 1 = 1/4$. So, the probability of getting a point in this interior region is $0.2 * 1/4 = 0.05$.

$P(X \le 1/4) = 0.1 + 0.4 + 0.05 = 0.55$

*b) $P(X + Y \le 1)$*

The corner points meeting this criterion are $(0, 0)$ ($P = 0.1$), $(1, 0)$ ($P = 0.2$), and $(0, 1)$ ($P = 0.4$). The area of the interior region meeting this criterion is the area of a triangle with base 1 and height 1, which is $1/2$. So, the probability of getting a point in this interior region is $0.2 * 1/2 = 0.1$.

$P(X + Y \le 1) = 0.1 + 0.2 + 0.4 + 0.1 = 0.8$

$$
f(x, y) =

    k & \text{for } a \le x \le b \text{ and } c \le y \le d \\
    0 & \text{otherwise}

$$

The support will be a rectangle. $X$ and $Y$ are independent, and the marginal distribution of each is uniformly distributed.

*b) Are X and Y independent?*

The distributions are independent if $P(X)P(Y) = P(X, Y)$.

7. **Suppose that the joint p.d.f. of X and Y is as follows. Are $x$ and $y$ independent?**

$$
f(x, y) =
\begin{cases}
    2xe^{-y} & \text{for } 0 \le x \le 1 \text{ and } 0 \lt y \lt \infty \\
    0 & \text{otherwise}
\end{cases}
$$



Find the marginal distribution of $Y$ for $y$ within bounds.

$\int f(x, y) dx = \int 2xe^{-y} dx = x^2 e^{-y}$

$P(Y) = \int_0^1 P(X, Y) dx = e^{-y}$


Find the marginal distribution of $X$ for $x$ within bounds.

$\int f(x, y) dy = \int 2xe^{-y} dy = -2xe^{-y}$

$P(X) = \int_0^\infty P(X, Y) dy$

$= \left( \lim_{y \to \infty} -2xe^{-y} \right) - -2xe^{-0}$

$= 2x$

$P(X)P(Y) = 2x e^{-y} = P(X, Y)$

So, $X$ and $Y$ are independent.

$P(\text{senior} | \text{more than once}) = {P(\text{senior} \land \text{more than once}) \over P(\text{more than once})}$

12. **Let Y be the rate (calls per hour) at which calls arrive at a switchboard. Let X be the number of calls during a two-hour period. Suppose that the marginal p.d.f. of Y is**

$$
f_2(y) =
\begin{cases}
    e^{-y} & \text{for } y > 0 \\
    0 & \text{otherwise}
\end{cases}
$$

and that the conditional p.f. of X given Y= y is

$$
g_1(x|y) =
\begin{cases}
    {(2y)^x \over x!}e^{-2y} & \text{for } x = 0, 1, 2, \ldots, y \\
    0 & \text{otherwise}
\end{cases}
$$


*a) Find the marginal p.f. of X using $\int_0^\infty y^k e^{-y}dy = k!$*

For $y \le 0$ or $x$ which is not a non-negative integer, $P(X, Y) = 0$.

For $y > 0$ and $x$ a non-negative integer,

$P(X, Y) = P(X | Y)P(Y) = \left( (x!)^{-1} (2y)^x e^{-2y} \right) \left( e^{-y} \right)$

We obtain the marginal distribution of $X$,

$P(X) = \int_0^\infty P(X, Y) dy$

$= \int_0^\infty \left( (x!)^{-1} (2y)^x e^{-2y} \right) \left( e^{-y} \right) dy$

$= (x!)^{-1} \int_0^\infty (2y)^x e^{-3y}dy$

$= (x!)^{-1} 2^x \int_0^\infty y^x e^{-3y}dy$

Let $u = 3y$, so $du = 3dy$. Then, the above becomes,

$= (x!)^{-1} 2^x \int_0^\infty ({u \over 3})^x e^{-u}{1\over 3}du$

$= (x!)^{-1} 2^x ({1 \over 3})^{x + 1} \int_0^\infty u^x  e^{-u}du$

Using the provided formula, we obtain,

$= (x!)^{-1} 2^x ({1 \over 3})^{x + 1} x!$

$P(X) = 2^x ({1 \over 3})^{x + 1}$

*b) Find the conditional p.d.f. $g_2(y | 0)$ of $Y$ given $X = 0$*

$P(Y | X) = {P(X, Y) \over P(X)}$

For $y \gt 0$ and $x$ a non-negative integer,

$= {\left( {(2y)^x e^{-3y} \over x!} \right) \over 2^x ({1 \over 3})^{x + 1}}$

$= {y^x e^{-3y} \over x! ({1 \over 3})^{x + 1}}$

So,

$P(Y | X = 0) = {y^0 e^{-3y} \over 0! ({1 \over 3})^{0 + 1}}$

$= {e^{-3y} \over {1 \over 3}}$

$= 3e^{-3y}$

*c) Find the conditional p.d.f. $g_2(y | 1)$ of $Y$ given $X = 1$*

$P(Y | X = 1) = {y^1 e^{-3y} \over 1! ({1 \over 3})^{1 + 1}}$

$= {y e^{-3y} \over {1 \over 9}}$

$= 9y e^{-3y}$

*d) For what values of $y$ is $g_2(y | 1) > g_2(y | 0)$? Does this agree with the intuition that the more calls you see, the higher you should think the rate is?*

```python
y = np.linspace(0, 10, 100)
f1 = 9 * y * np.exp(-3 * y)
f0 = 3 * np.exp(-3 * y)

plt.plot(y, f1, label="P(Y | X = 1)")
plt.plot(y, f0, label="P(Y | X = 0)")
plt.ylabel("probability density")
plt.xlabel("Y (calls per hour)")
plt.legend()
plt.show()
```

$P(Y | X = 1) > P(Y | X = 0)$

$9y e^{-3y} > 3e^{-3y}$

$9y > 3$

$y > {1 \over 3}$

This is saying that, for any rate of calls $y > {1 \over 3}$, this rate of calls is more likely if you observed one call than if you observed no calls. So, yes, higher rate is consistent with higher number of calls.
