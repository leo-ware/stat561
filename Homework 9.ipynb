5. A coin for which the probability of heads is 0.6 is tossed
nine times. Use the table of the binomial distribution given
at the end of this book to find the probability of obtaining
an even number of heads.



$P(H) = 0.6$ is the same as $P(T) = 0.4$. There will be an even number of heads iff there is an odd number of tails.

| # tails | Probability |
| - | - |
| 1 | 0.0605 |
| 3 | 0.2508 |
| 5 | 0.1672 |
| 7 | 0.0212 |
| 9 | 0.0003 |

These are mutually exclusive, so we just add them up.

$0.0605 + 0.2508 + 0.1672 + 0.0212 + 0.0003 = 0.5$


6. Three men A, B, and C shoot at a target. Suppose that
A shoots three times and the probability that he will hit
the target on any given shot is 1/8, B shoots five times and
the probability that he will hit the target on any given shot
is 1/4, and C shoots twice and the probability that he will
hit the target on any given shot is 1/2. What is the expected
number of times that the target will be hit?

The EV of a binomial distribution is $np$. These are independent, so we just add them up.

$3 \cdot \frac{1}{8} + 5 \cdot \frac{1}{4} + 2 \cdot \frac{1}{2} = \frac{3}{8} + \frac{5}{4} + 1 = \frac{3}{8} + \frac{10}{8} + \frac{8}{8} = \frac{21}{8} = 2.625$

4. If a random variable X has the hypergeometric distribution with parameters A= 8, B= 20, and n, for what
value of n will Var(X) be a maximum?

$\text{Var}(X) = {nAB(A+B-n) \over (A+B)^2(A+B-1)}$

We can drop the denominator because it is constant with respect $n$.

$\text{Var}(X) \propto nAB(A+B-n) = nA^2B + nAB^2-n^2AB$

Call this function $f(n)$.

$f(n) = n(A^2B + AB^2-nAB)$

Take the derivative with respect to $n$.

${d \over dn} f(X) = {d \over dn} (nA^2B + nAB^2-n^2AB) = A^2B + AB^2-2nAB$

Set the derivative equal to 0 and solve for $n$.

$A^2B + AB^2 - 2nAB = 0$

$n = {A^2B + AB^2 \over 2AB} = {A + B \over 2}$

So, variance is maximized when $n = {A + B \over 2}$, the mean value of $A$ and $B$.


This would be great is $n$ were real valued, but it's actually an integer. So, when $A + B$ is odd, ${A + B \over 2}$ doesn't work. I claim that the floor and ceiling of this value maximize variance in this case.

To prove this, let's consider what happens to $f(n)$ for values near ${A + B \over 2}$.

$f({A + B \over 2} + d)$

$= A^2B + AB^2 - 2({A + B \over 2} + d)AB$

$= A^2B + AB^2 - (A + B + 2d)AB$

$= A^2B + AB^2 - (A^2B + AB^2 + 2dAB)$

$= -2dAB$

$A$ and $B$ are positive, so $-2dAB$ positive for negative $d$ and negative for positive $d$. So, as $n$ moves away from ${A + B \over 2}$, the variance decreases. And, because this function is linear in $d$, we know that it's decreasing at the same rate on both sides of ${A + B \over 2}$. So, the maximum at ${A + B \over 2}$ is unique, and $\text{Var}({A + B \over 2} + d) = \text{Var}({A + B \over 2} - d)$ for any $d$.

The integer value of $n$ that minimizes variance is the integer which is closest to ${A + B \over 2}$. If $A + B$ is odd, then the floor and ceiling of ${A + B \over 2}$ are both 0.5 from it. So, both $n = \text{floor}({A + B \over 2})$ and $n = \text{ceil}({A + B \over 2})$ maximize variance.

We use this formula to solve for the specific case where $A = 8$ and $B = 20$. The answer is ${20 + 8 \over 2} = 14$.

2. Suppose that on a given weekend the number of accidents at a certain intersection has the Poisson distribution
with mean 0.7. What is the probability that there will be at least three accidents at the intersection during the weekend?

The poisson distribution is given by, $P(k | \lambda) = {\lambda^k e^{-\lambda} \over k!}$

$P(k \geq 3 | \lambda = 0.7)$

$= 1 - P(k < 3)$

$= 1 - (P(k = 0) + P(k = 1) + P(k = 2))$

$= 1 - e^{-0.7}\left( {0.7^0  \over 0!} + {0.7^1 \over 1!} + {0.7^2 \over 2!} \right)$

$= 1 - e^{-0.7}\left( 1 + 0.7 + {0.49 \over 2} \right)

$= 1 - e^{-0.7}1.945$

$= 1 - 0.9658584158742916$

$= 0.034141584125708446$

2. Suppose that a sequence of independent tosses are
made with a coin for which the probability of obtaining a
head on each given toss is 1/30.
    - a. What is the expected number of tails that will be
    obtained before five heads have been obtained?
    - b. What is the variance of the number of tails that will
    be obtained before five heads have been obtained?

We can model this scenario with a negative binomial distribution with parameters $r = 5$ and $p = 1/30$. Then, we can use the formulas for the expected value and variance of a negative binomial distribution.

$E[X] = r{1-p \over p} = 5{29/30 \over 1/30} = 5*29 = 145$

$Var(X) = r{1-p \over p^2} = 5{29/30 \over (1/30)^2} = 5*29*30 = 4350$

6. Suppose that X has the geometric distribution with
parameter p. Determine the probability that the value of
X will be one of the even integers 0, 2, 4, . .

The pmf of the geometric distribution is $P(k | p) = p(1-p)^k$., so we need the sum of $P(2k | p)$ for nonnegative integer $k$.

$\sum_{k=0}^{\infty} p(1-p)^{2k}$

$= p\sum_{k=0}^{\infty} (1-p)^{2k}$

$= p\sum_{k=0}^{\infty} ((1-p)^{2})^k$

We use the formula for the sum of a geometric series.

$= {p \over 1 - (1-p)^2}$

$= {p \over 1 - (p^2 - 2p + 1)}$

$= {p \over 2p - p^2}$

$= {1 \over 2 - p}$
